```python
import os
from dotenv import load_dotenv
from IPython.display import Markdown,display
from openai import OpenAI


openai = OpenAI()

load_dotenv(override = True)
api_key = os.getenv('OPENAI_API_KEY')

if not api_key:
    print("No API key found")
elif not api_key.startswith("sk-proj-"):
    print("Invalid API key")
elif not api_key.strip() == api_key:
    print("API key has extra spaces remove them")
else:
    print("API key is valid")


system_prompt = """
You are a helpful assistant that analyses the content of an email and gives an appropriate formal Subject for it"
"""
user_prompt_prefix = """
These are the contents of the email, analyse the whole text
"""

def messages(email):
    return [
        {"role":"system","content":system_prompt},
        {"role":"user","content":user_prompt_prefix + email}
    ]

def subject_generator(email):    
    response = openai.chat.completions.create(model="gpt-4.1-nano", messages= messages(email))
    return "Subject: " + response.choices[0].message.content

    
email = """ 
I hope this email finds you well and in good spirits. It’s been a while since we last connected, so I thought I’d take a moment to reach out and share a few updates, along with some general thoughts that have been on my mind lately.

Over the past few weeks, things have been quite busy on my end, but in a good way. I’ve been juggling multiple tasks, exploring new ideas, and trying to bring a bit more structure into my daily routine. It’s interesting how time seems to move faster when you’re fully immersed in work, yet slows down when you finally get a moment to reflect and breathe.

I’ve also been thinking about how important it is to maintain meaningful connections, even in the middle of hectic schedules and endless responsibilities. Sometimes, a simple message or a quick conversation can make a significant difference in someone’s day. That’s one of the reasons I wanted to write this email — not for any urgent reason, but just to stay in touch and keep the communication flowing.

On another note, I’ve been working on improving my productivity and exploring ways to stay more organized. From experimenting with new planning methods to refining my workflow, it’s been a process of learning and adapting. It’s surprising how small changes, like planning the day the night before or setting aside distraction-free time, can have such a positive impact.

If there’s anything new happening on your side, I’d genuinely love to hear about it. Whether it’s related to work, personal projects, or just life in general, feel free to share. It’s always refreshing to exchange ideas and updates, even if it’s just a short message.

Thank you for taking the time to read this rather lengthy email. I truly appreciate it. Let’s not let too much time pass before our next conversation.

Wishing you a wonderful day ahead and looking forward to hearing from you soon.

Warm regards,
Vinay
"""

subject_generator(email)




```
